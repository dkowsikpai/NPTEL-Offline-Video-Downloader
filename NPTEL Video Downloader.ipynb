# NPTEL Video Offline Downloader

```python
# Import Libraries
from bs4 import BeautifulSoup # pip install beautifulsoup4
from bs4.dammit import EncodingDetector
from requests import get
```

```python
# Setting parser and fetching data
parser = 'html.parser'  # or 'lxml' (preferred) or 'html5lib', if installed
resp = get("https://nptel.ac.in/courses/106/106/106106169/")
```

```python
# Encoding
http_encoding = resp.encoding if 'charset' in resp.headers.get('content-type', '').lower() else None
html_encoding = EncodingDetector.find_declared_encoding(resp.content, is_html=True)
encoding = html_encoding or http_encoding
soup = BeautifulSoup(resp.content, parser, from_encoding=encoding)
```

```python
soup
```

```python
# https://nptel.ac.in/content/storage2/106/106/106106169/MP4/mod01lec01.mp4
l = []
x = soup.find('table', {'id': "request"})
for i in x.find_all("tr"):
    tds = i.find_all("td")
    if len(tds) == 3:
        a = tds[2].find("a")
        href = a["href"]
        if href[-3:] == "mp4":
            l.append({
                "name": tds[1].text + ".mp4",
                "link": "https://nptel.ac.in" + href
            })
print(l)
```

```python
def download(url, name):
    r = get(url) # create HTTP response object 
    # send a HTTP request to the server and save 
    # the HTTP response in a response object called r 
    with open("./videos/"+name, 'wb') as f: 
        f.write(r.content) 
```

```python
download(l[0]["link"], l[0]["name"])
```

```python
for i in range(1, 13+1):
    print("Downloading", l[i-1]["name"])
    download(l[i-1]["link"], l[i-1]["name"])
    print("Downloaded")
```
